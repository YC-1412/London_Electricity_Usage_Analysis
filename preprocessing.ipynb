```python
import pandas as pd
import numpy as np
from pandas import datetime
import os
import math
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import calendar
sns.set(color_codes=True)
```

```python
acorn = pd.read_csv('informations_households.csv')
acorn.head()
```

```python
daily_acorn = pd.merge(daily[['LCLid', 'day','energy_sum']],acorn[['LCLid','Acorn']], on='LCLid')#,how='left')
```

```python
daily_acorn['date'] = pd.to_datetime(daily_acorn['day'])
```

```python
daily_acorn['day_of_week'] = daily_acorn['date'].apply(lambda x: x.weekday()) # get the weekday index, between 0 and 6
daily_acorn['day_of_week'] = daily_acorn['day_of_week'].apply(lambda x: calendar.day_name[x])
daily_acorn.head()
```

```python
daily_acorn['weekend']= pd.Series(np.where((daily_acorn.day_of_week.values == 'Saturday') | (daily_acorn.day_of_week.values == 'Sunday'), 1, 0),daily_acorn.index)
daily_acorn.head()
```

```python
daily_acorn.day.value_counts()
```

```python
holidays_df=pd.read_csv("UK Bank Holidays.csv")
print(holidays_df.shape)
holidays_df.head()
```

```python
holidays_df=holidays_df[holidays_df['Bank Holiday']==1].reset_index(drop=True)
holidays_df.head()
```

```python
holidays=holidays_df['Date']
daily_acorn['Holiday'] = daily_acorn['day'].isin(holidays)
daily_acorn['Holiday'] =daily_acorn['Holiday'].astype(int)
daily_acorn.sample(20)
```

```python
daily_acorn['month'] = daily_acorn.date.dt.month
daily_acorn['year'] = daily_acorn.date.dt.year
```

```python
sp = list(np.where((daily_acorn.month.values == 3) | (daily_acorn.month.values == 4) | (daily_acorn.month.values == 5))[0])
su = list(np.where((daily_acorn.month.values == 6) | (daily_acorn.month.values == 7) | (daily_acorn.month.values == 8))[0])
au = list(np.where((daily_acorn.month.values == 9) | (daily_acorn.month.values == 10) | (daily_acorn.month.values == 11))[0])
wi = list(np.where((daily_acorn.month.values == 12) | (daily_acorn.month.values == 1) | (daily_acorn.month.values == 2))[0])
```

```python
daily_acorn['season'] = ''
daily_acorn.loc[sp, 'season'] = 'Spring'
daily_acorn.loc[su, 'season'] = 'Summer'
daily_acorn.loc[au, 'season'] = 'Autumn'
daily_acorn.loc[wi, 'season'] = 'Winter'
```

```python
daily_acorn=daily_acorn.drop(['date'],axis=1)
```
